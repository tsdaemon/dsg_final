```python
def get_cathegorical_stats(df, column_name):
    final_df = df.groupby(['SalOrg', 'Material', 'Month', column_name])['OrderQty']
    final_df_sum = final_df.sum().unstack(column_name).add_prefix(column_name + 'sum_')
    final_df_count = final_df.size().unstack(column_name).add_prefix(column_name + 'count_')
    final_df_mean = final_df.mean().unstack(column_name).add_prefix(column_name + 'mean_')
    final_df_max = final_df.max().unstack(column_name).add_prefix(column_name + 'max_')
    final_df_min = final_df.min().unstack(column_name).add_prefix(column_name + 'min_')
    final_df = final_df_sum.reset_index().merge(final_df_count.reset_index(), on = ['SalOrg', 'Material', 'Month'], copy = False)
    final_df = final_df.merge(final_df_mean.reset_index(), on = ['SalOrg', 'Material', 'Month'], copy = False)
    final_df = final_df.merge(final_df_max.reset_index(), on = ['SalOrg', 'Material', 'Month'], copy = False)
    final_df = final_df.merge(final_df_min.reset_index(), on = ['SalOrg', 'Material', 'Month'], copy = False)
    final_df = final_df.fillna(0)
    return final_df
```

```python
def get_clustered_stats(df, column_name):
    final_df = df.groupby(['SalOrg', 'Month', column_name])['OrderQty']
    final_df_sum = final_df.sum().unstack(column_name).add_prefix(column_name + 'sum_')
    final_df_count = final_df.size().unstack(column_name).add_prefix(column_name + 'count_')
    final_df_mean = final_df.mean().unstack(column_name).add_prefix(column_name + 'mean_')
    final_df_max = final_df.max().unstack(column_name).add_prefix(column_name + 'max_')
    final_df_min = final_df.min().unstack(column_name).add_prefix(column_name + 'min_')
    final_df = final_df_sum.reset_index().merge(final_df_count.reset_index(), on = ['SalOrg', column_name, 'Month'], copy = False)
    final_df = final_df.merge(final_df_mean.reset_index(), on = ['SalOrg', column_name, 'Month'], copy = False)
    final_df = final_df.merge(final_df_max.reset_index(), on = ['SalOrg', column_name, 'Month'], copy = False)
    final_df = final_df.merge(final_df_min.reset_index(), on = ['SalOrg', column_name, 'Month'], copy = False)
    final_df = final_df.fillna(0)
    return final_df
```

```python
def get_mean_stats(df, column_name):
    final_df = df.groupby(['SalOrg', 'Material', 'Month'])[column_name].mean().reset_index()
    return final_df
    
```

```python
from scipy.stats.stats import pearsonr   
def get_correlation(df, n_months, n_delay):
    
    train = df.groupby(['SalOrg', 'Month', 'Material'])['OrderQty'].apply(lambda x: x.sum()).reset_index()
    
    train['LaggedOrderQty'] = train.OrderQty.shift(12)
    train = train.fillna(0)
    train = train.sort_values(by=['SalOrg', 'Material','Month'])
    df_corr = train.groupby(['SalOrg', 'Material']).apply(lambda x: np.corrcoef(np.array(x['OrderQty']), np.array(x['LaggedOrderQty']))[0][1]).reset_index()
    df_corr.columns = ["SalOrg"] + ['Material'] + ['Corr']
    return df_corr
```
